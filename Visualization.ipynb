```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as pit

%matplotlib inline
pit.style.use("classic")
```

```python
x = np.linspace(0, 10, 100)

fig1 = pit.figure()
pit.plot(x, np.sin(x))
pit.plot(x, np.cos(x))

pit.show()

fig1.savefig("myfig.png")

fig2 = pit.figure()
pit.plot(x, np.sin(x))

pit.title("A Sine Curve")
pit.xlabel("x")
pit.ylabel("y")
```

```python
data = np.random.randn(1000)
fig3 = pit.figure()
pit.hist(data)
pit.hist(data, bins=30)
fig3.savefig("myhist.png")
pit.title("Histogram")
pit.xlabel("X")
pit.ylabel("Frequency")
```

```python
norm = np.random.standard_normal(1000)
fig4 = pit.figure()
pit.hist(norm, bins=50)
fig4.savefig("Normal.png")
pit.title("Standard Normal Distribution")
pit.xlabel("X")
pit.ylabel("Y")
```
